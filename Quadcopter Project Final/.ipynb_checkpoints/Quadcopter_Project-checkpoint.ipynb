## Controlling the Quadcopter

The provided sample agent can be run in the code cell below.

The agent controls the quadcopter by setting the revolutions per second on each of its four rotors.  The provided agent in the `Basic_Agent` class below always selects a random action for each of the four rotors.  These four speeds are returned by the `act` method as a list of four floating-point numbers.  

The agent I implemented in `agents/agent.py` has a a far more intelligent method for selecting actions.

```python
import random

class Basic_Agent():
    def __init__(self, task):
        self.task = task
    
    def act(self):
        new_thrust = random.gauss(450., 25.)
        return [new_thrust + random.gauss(0., 1.) for x in range(4)]
```

In the code cell below, the agents select actions to control the quadcopter.  

The `labels` list below annotates statistics that are saved while running the simulation.  All of this information is saved in a text file `data.txt` and stored in the dictionary `results`.  

```python
%load_ext autoreload
%autoreload 2

import csv
import numpy as np
from task import Task

# Modify the values below to give the quadcopter a different starting position.
runtime = 5.                                     # time limit of the episode
init_pose = np.array([0., 0., 10., 0., 0., 0.])  # initial pose
init_velocities = np.array([0., 0., 0.])         # initial velocities
init_angle_velocities = np.array([0., 0., 0.])   # initial angle velocities
file_output = 'data.txt'                         # file name for saved results

# Setup
task = Task(init_pose, init_velocities, init_angle_velocities, runtime)
agent = Basic_Agent(task)
done = False
labels = ['time', 'x', 'y', 'z', 'phi', 'theta', 'psi', 'x_velocity',
          'y_velocity', 'z_velocity', 'phi_velocity', 'theta_velocity',
          'psi_velocity', 'rotor_speed1', 'rotor_speed2', 'rotor_speed3', 'rotor_speed4']
results = {x : [] for x in labels}

# Run the simulation, and save the results.
with open(file_output, 'w') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(labels)
    while True:
        rotor_speeds = agent.act()
        _, _, done = task.step(rotor_speeds)
        to_write = [task.sim.time] + list(task.sim.pose) + list(task.sim.v) + list(task.sim.angular_v) + list(rotor_speeds)
        for ii in range(len(labels)):
            results[labels[ii]].append(to_write[ii])
        writer.writerow(to_write)
        if done:
            break
```



## The Task

The code for my task can be found in `task.py`.  

The `__init__()` method is used to initialize several variables that are needed to specify the task.  

The `reset()` method resets the simulator.  The agent calls this method every time the episode ends.  You can see an example of this in the code cell below.

The `step()` method accepts the agent's choice of action `rotor_speeds`, which is used to prepare the next state to pass on to the agent.  Then, the reward is computed from `get_reward()`.  The episode is considered done if the time limit has been exceeded, or the quadcopter has travelled outside of the bounds of the simulation.

## The Agent

The sample agent given in `agents/policy_search.py` uses a very simplistic linear policy to directly compute the action vector as a dot product of the state vector and a matrix of weights. Then, it randomly perturbs the parameters by adding some Gaussian noise, to produce a different policy. Based on the average reward obtained in each episode (`score`), it keeps track of the best set of parameters found so far, how the score is changing, and accordingly tweaks a scaling factor to widen or tighten the noise.

The code cell below shows how the agent performs on the sample task.

```python
import sys
import pandas as pd
from agents.policy_search import PolicySearch_Agent
from task import Task

num_episodes = 100
target_pos = np.array([0., 0., 10.])
task = Task(target_pos=target_pos)
agent = PolicySearch_Agent(task) 

for i_episode in range(1, num_episodes+1):
    state = agent.reset_episode() # start a new episode
    while True:
        action = agent.act(state) 
        next_state, reward, done = task.step(action)
        agent.step(reward, done)
        state = next_state
        if done:
            print(i_episode, agent.score, agent.best_score, agent.noise_scale)
            print("\rEpisode = {:4d}, score = {:7.3f} (best = {:7.3f}), noise_scale = {}".format(
                i_episode, agent.score, agent.best_score, agent.noise_scale), end="")  # [debug]
            break
    sys.stdout.flush()
```

This provided agent performs very poorly on this task.  

## Implementation: Define the Task, Design the Agent, and Train Agent

I amended `task.py` to specify a task of my choosing.  

I defined my own agent in `agents/agent.py`. 

```python
## I trained my agent here.
import sys
import pandas as pd
from agents.agent import DDPG_Agent
from task import Task

num_episodes = 1500
target_pos = np.array([0., 0., 10.])
task = Task(target_pos=target_pos)
agent = DDPG_Agent(task) 
current_best = -np.inf

allRewards = []
allClose = [] #all the times the agent got close to the target height in the episode
allBestRewards = [current_best]

for i_episode in range(1, num_episodes+1):
    state = agent.reset_episode() # start a new episode
    while True:
        action = agent.act(state) 
        next_state, reward, done = task.step(action)
        agent.step(action,reward, next_state, done)
        state = next_state
        if done:
            allRewards.append(agent.total_reward)
            allClose.append(agent.closeCount)
            if(agent.best_score > current_best):
                allBestRewards.append(agent.best_score)
                current_best = agent.best_score
            print("\rEpisode = {:4d}, best_score = {:7.3f}, total_reward = {:7.3f}".format(
                i_episode, agent.best_score, agent.total_reward), end="") # [debug]
            print(task.sim.pose[:3])
            #print("\rEpisode = {:4d}, score = {:7.3f} (best = {:7.3f}), noise_scale = {}".format(
            #    i_episode, agent.score, agent.best_score, agent.noise_scale), end="")  # [debug]
            break
    sys.stdout.flush()
print(allClose)
```

## Reflections

**Question 1**: Describe the task that you specified in `task.py`.  How did you design the reward function?


**Answer:** 
I decided to use the example task of moving to a target position. A Udacity reviewer suggested I focus on the Z target value only. Although the ending position of my quadcopter was often 0 (if crashed) or too high, I decided to plot out how many steps per episode the quadcopter spent close to the target height. For the value of what is considered "close" I used + or - 1/2 a unit from the target Z value. As can be seen in the chart above, the quadcopter actually does spend a good amount of time close to the target Z value and starts doing that after about 100 episodes. It has a hard time staying there, but it did figure out reaching there (which is what I had set for my task) fairly quickly. The reward function I used is: 

- 1) a base value of reward = 1- .003*(abs(self.sim.pose[:1] - self.target_pos[2]).sum())
- 2) As suggested by the reviewer, I incorporated the vertical velocity into the reward function. to encourage flying when too low, if the current position is below the target position, reward += np.clip(self.sim.v[2], -.1, .1)
- 3) to discourage flying too high, if the current position is twice as high as the current position, reward -= np.clip(self.sim.v[2], -.1, .1)
- 4) this is then normalized, reward = np.tanh(reward)
- 5) a large bonus is given for being close to the target. Reward + 1.5 if within 1 unit, reward + 3.5 if within half a unit.

While my rewards chart doesnt have a nice gradual rise, it actually performs well overall. During testing, I also realized that my task was not just "reach Z height." It was "reach Z height and stay there until the end of the episode." which it actually managed to do fairly well as described below, even if the position at the very end of the episode wasn't always perfect. 


**Question 2**: Discuss your agent briefly, using the following questions as a guide:

- What learning algorithm(s) did you try? What worked best for you?
- What was your final choice of hyperparameters (such as $\alpha$, $\gamma$, $\epsilon$, etc.)?
- What neural network architecture did you use (if any)? Specify layers, sizes, activation functions, etc.

**Answer:** 
1) I used Deep Deterministic Policy Gradient (DDPG) for my agent. I thought this would work well due to the continuous nature of the action space. My project uses the idea of experience replay. Experiences are stored in a fixed size memory buffer (10000 in my case) and sampled in random batches to learn from so that previous experiences can be learned from and there is no bias from sequential entries. I changed the batch size from 1024 in my first submission to 64 in this one. I did not see a decrease in performance, I actually saw an increase, though that may be the change in my reward function. 

2) My final choice of hyperparameters were: 
- buffer_size = 100000
- batch_size = 1024
- gamma = 0.99  
- tau = 0.01  
- L2 regularization value = .000001
- exploration mu = 0
- exploration theta = 0.15
- exploration sigma = 0.2

I used the idea of action repeats in my project. For each step of the agent, the simulation in my project steps 6 times. With this information of changes through time, the agent can infer velocity. 

3) The neural network architecture I used was:
Actor Neural Network:
Dense layer with 64 units, relu activation function, L2 regularization, and Batch Normalization
Dense layer with 128 units, relu activation function, L2 regularization, and Batch Normalization
Final output layer with sigmoid activation function

Critic States Neural Network:
Dense layer with 64 units, relu activation function, L2 regularization, and Batch Normalization
Dense layer with 128 units, relu activation function, L2 regularization, and Batch Normalization

Critic Actions Neural Network:
Dense layer with 128 units, relu activation function, L2 regularization, and Batch Normalization

Critic Final Layer:
Final output layer combine States and Actions with sigmoid activation

I added an implementation of OrnsteinUhlenbeck Noise into my project. We want to introduce some noise so that different actions are explored and it is like sampling from a Gaussian distribution. However we also want consecutive samples to be not too far of from each other so that the quadcopter has smooth transitions instead of jerking around everywhere.
